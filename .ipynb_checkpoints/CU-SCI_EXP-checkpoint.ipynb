```python
from cunet_model import CUNet
from cunet_parts import *
%run functions_exp.ipynb

dtype = torch.cuda.FloatTensor
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# setup parameters

LR =  0.05  # Learning rate of deep neural networks
num_iter = 301  # # Total number of iterations
path = "11-11-24_06-18N64-P384-S6000-11.47h-Random-SR10-Seed123-data.csv" 
rate = 0.5  # sampling rate

show_every = 100
seed = 123

# Data preparation

shape = [64,64]
num = int(shape[0] * shape[1] * rate)
real,imag = get_signal(path,num)

real = real.type(dtype).detach()
imag = imag.type(dtype).detach()
real = real.to(device)
imag = imag.to(device)

rng = np.random.default_rng(seed)
p = get_random_pattern(rng,num,shape)
p = p.type(dtype).detach()
p = p.to(device)

# Model preparation

in_channels = 1
out_channels = 1
model = CUNet(in_channels, out_channels)
model = model.to(device)

rand_input_real = torch.rand(1, 1, 64, 64)
rand_input_imag = torch.rand(1, 1, 64, 64)
rand_input = torch.complex(rand_input_real, rand_input_imag).to(device)

optimizer = torch.optim.Adam(params=model.parameters(), lr=LR)

# Image reconstruction

start_time = time.time()

for i in range(num_iter):
    pred_img = model(rand_input)
    final_real = torch.real(pred_img)
    final_imag = torch.imag(pred_img)
    
    real_pred = get_signal1(final_real, p)
    imag_pred = get_signal2(final_imag, p)
    mse = torch.nn.MSELoss()
    loss = mse(real,real_pred) + mse(imag,imag_pred)
        
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

end_time = time.time()

execution_time = end_time - start_time
print("Execution time: ", execution_time, "seconds")

phase_shift(final_real,final_imag)
```
