```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

file_name = "11-11-24_06-18N64-P384-S6000-11.47h-Random-SR10-Seed123-data.csv"
samp_ratio = 1
d = 4
phi = 1.0

mdata = pd.read_csv(file_name, header=None, names=["No.", "Value"])
mear_data = mdata["Value"].values.reshape(-1, 1)
random_seed = 123
rng = np.random.default_rng(random_seed)
N= 64
th = 0.5
fram_num = int(samp_ratio * N **2)
num_chans = 1

def RandomRecover4PM(rng, mear_data, M, N, Num, th=0.5, num_chans=1):

    mdata_re = mear_data[0::4, :] - mear_data[1::4, :]
    mdata_im = mear_data[2::4, :] - mear_data[3::4, :]

    rec_image_re = np.zeros((M, N, num_chans))
    rec_image_im = np.zeros((M, N, num_chans))

    for colorch in range(num_chans):
        cimage_re = np.zeros((M, N))
        cimage_im = np.zeros((M, N))

        for n in range(Num):
            P = rng.random((M, N))
            P_pos = (P > th).astype(np.float64)
            P_neg = 1 - P_pos
            P = P_pos - P_neg

            cimage_re += mdata_re[n, colorch] * P
            cimage_im += mdata_im[n, colorch] * P

        rec_image_re[:, :, colorch] = cimage_re
        rec_image_im[:, :, colorch] = cimage_im

    rec_image_re = rec_image_re.squeeze()
    rec_image_im = rec_image_im.squeeze()

    return rec_image_re, rec_image_im

rec_image_re, rec_image_im = RandomRecover4PM(rng, mear_data, N, N, fram_num, th, num_chans)

fn = rec_image_re + 1j * rec_image_im
fn = fn * np.exp(1j * phi)
amplitude__ = np.abs(fn)
phase__ = np.angle(fn)

plt.figure(figsize=(d, d))
plt.imshow(rec_image_re)
plt.axis('off')
plt.show()

plt.figure(figsize=(d, d))
plt.imshow(rec_image_im)
plt.axis('off')
plt.show()

plt.figure(figsize=(d, d))
plt.imshow(amplitude__)
plt.axis('off')
plt.show()

plt.figure(figsize=(d, d))
plt.imshow(phase__)
plt.axis('off')
plt.show()
```
